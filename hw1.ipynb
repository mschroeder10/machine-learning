```python
df = pd.read_csv("kc_house_data.csv")
df = df.drop(columns = ["id", "date", "zipcode"])
```

```python
#5a
# calculate average
print("Feature Average:")
mean = df.mean()
print(mean, "\n")

# calculate max
print("Feature Max")
print(df.max(), "\n")

# calculate min
print("Feature Min")
print(df.max(), "\n")

# calculate variance
print("Feature Variance")
variance = df.var()
print(variance, "\n")
```

```python
print("Highest Average:")
print (pd.DataFrame({'feature':mean.index, 'value':mean.values}).max(), "\n")
print ("Lowest Average:")
print (pd.DataFrame({'feature':mean.index, 'value':mean.values}).min(), "\n")

print("Highest Variance:")
print (pd.DataFrame({'feature':variance.index, 'value':variance.values}).max(), "\n")
print ("Lowest Variance:")
print (pd.DataFrame({'feature':variance.index, 'value':variance.values}).min(), "\n")
```

```python
#5b
max_corr = -10000
max_feature_row = ""
max_feature_col = ""
corr = df.corr()
for col in corr:
    drop_self = corr[col].drop(index = col)
    temp = pd.DataFrame({'feature':drop_self.index, 'correlation':drop_self.values})
    max_temp_corr = temp.max()
    if (max_temp_corr['correlation'] > max_corr):
        max_corr = max_temp_corr['correlation']
        max_feature_row = max_temp_corr['feature']
        max_feature_col = col
print ("features with highest correlation coefficient: ", max_feature_row, " ", max_feature_col)
print (max_corr)
```
